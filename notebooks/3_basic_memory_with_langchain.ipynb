```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model ="gemini-2.5-flash-lite")
```

```python
prompts = [
    {"role": "user", "content": "Hello, My name is Tamanna"},
    {"role": "ai", "content": "Hello Tamanna, how can I help you today?"},
    {"role": "user", "content": "Can you remind me what my name is?"}
]

res = llm.invoke(prompts)
```

```python
res.content
```

#### Manage History

```python
history = []    
while True:
    query = input("user: Enter exit or quit to end chat ")
    if query.lower() in ["exit", "quit"]:
        print("Exiting...")
        break
    history.append({"role": "user", "content": query})
    print("User: ", query, "\n")

    res = llm.invoke(history)
    history.append({"role": "ai", "content": res.content})

    print("ai: ", res.content, "\n")
```
