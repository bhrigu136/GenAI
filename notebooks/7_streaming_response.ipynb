```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="openai/gpt-oss-20b", streaming=True)
```

```python
question = "Can you explain me GEN-AI ?"
```

```python
res = llm.stream(question)
for chunk in res:
    print(chunk.content, end="")
```
